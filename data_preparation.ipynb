<!-- 4. Data Cleaning -->


<!-- Menghapus Duplikat -->

```python
df = df.drop_duplicates()
print("\n=== Data Setelah Dihapus Duplikat ===")
print(df)
```

<!-- Missing Value -->

```python
# df = pd.read_csv("dataset_contoh.csv")
# print("\n=== Data Sebelum Menangani Nilai Hilang ===")
# print(df)

df['Age'] = df['Age'].fillna(df['Age'].median())
df['Salary'] = df['Salary'].fillna(df['Salary'].mean())
df['Department'] = df['Department'].fillna(df['Department'].mode()[0])
print("\n=== Data Setelah Menangani Nilai Hilang ===")
print(df)
```
